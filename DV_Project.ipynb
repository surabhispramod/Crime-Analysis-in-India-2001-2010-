# Analysis of crime cases reported against women in North India & South India

```python
ra=df[df['Area_Name']=='Karnataka']
ra_tot=ra.groupby('Year').sum()
r=ra_tot['Cases_Reported']
graph_compare=plt.figure(figsize=(12,8))
r_g=graph_compare.add_subplot(211)
r_g.set_xlabel('Years')
r_g.set_ylabel('Case Rate')
r_g.set_title("Crime cases reported against Women in Karnatka (South India)")
r_g.plot(r)
```

```python
na=df[df['Area_Name']=='Uttar Pradesh']
na_tot=na.groupby('Year').sum()
n=na_tot['Cases_Reported']
graph_compare=plt.figure(figsize=(12,8))
n_g=graph_compare.add_subplot(211)
n_g.set_xlabel('Years')
n_g.set_ylabel('Case Rate')
n_g.set_title("Crime cases reported against Women in West Bengal (North India)")
n_g.plot(n)
```

# Rape and Kindnapping & abduction reported in Bihar (2001-2010)

# Percentage of trials for the cases reported against women (2001-2010)

```python
total=df['Total_Cases_for_Trial']
sent=df['Cases_Sent_for_Trial']
completed=df['Cases_Trials_Completed']
colors = ("cyan", "brown","grey")
explode = (0.3, 0.1, 0.1)
plt.figure(figsize=(6,6))
plt.pie([total.sum(),sent.sum(),completed.sum()],labels=['Total_Cases_for_Trial','Cases_Sent_for_Trial','Cases_Trials_Completed'],
        autopct="%0.2f%%",explode=explode,shadow=True,startangle=90,colors=colors)
plt.title("Percentage of Trials for the cases reported between 2001 to 2010")
plt.show()
```

# Total crime cases against children in the state with total count of cases in district

```python
# Total cases against children in the state with the total count of cases in there districts(2005)
da=data[(data['Year']==2005)&(data['DISTRICT']=='TOTAL')]
d1=da['Total'].plot(kind='barh',figsize=(20,9),color='orange',alpha=0.65)
d1.set_title("Total cases against children statewise along with there districts(2005)")
d1.set_xlabel("Total Rate")
d1.set_ylabel("State /Ut")
d1.set_yticklabels(da['STATE/UT'])
plt.grid()
```

# Analysing an 3 crime cases againt children (2001 -2010)

```python
data.groupby("Year")[['Rape','Kidnapping and Abduction','Buying of girls for prostitution']].max().plot(kind='line',figsize=(22,7))
plt.title("Analysis of 3 crime cases against children from 2001 to 2010(India)")
```

# Percentage of crime against childern over the year 2001 to 2010

```python
rape=data['Rape']
KidnappingandAbduction=data['Kidnapping and Abduction']
other=data['Other Crimes']
murder=data['Murder']
colors = ['pink','cyan','grey','red']
explode = (0.2, 0.1, 0.3,0.1)
plt.figure(figsize=(6,6))
plt.pie([rape.sum(),KidnappingandAbduction.sum(),other.sum(),murder.sum()],labels=['Rape','Kidnapping and Abduction','Other Crimes','Murder'],
        autopct="%0.2f%%",startangle=85, colors = colors,explode=explode,shadow=True)
plt.title("Percentage of crime against childern over the year 2001 to 2010")
plt.show()
```
